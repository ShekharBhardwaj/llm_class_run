# 📚 Final Summary

---

## 🔹 Transformers

- Transformers are **the backbone of modern AI models** like GPT and Claude.
- They work by letting every word **pay attention to every other word** — instantly.
- **No step-by-step reading** like old models — Transformers **see the whole sentence at once**.

✅ **One-liner:**  
> **Transformers changed AI by using attention, not memory.**

---

## 🔹 Tokenization

- **Tokenization** is how text is **chopped into smaller units** for the model to process.
- Tokens can be:
  - **Whole words** (e.g., "cat")
  - **Parts of words** (e.g., "hand-crafted" → "hand" + "crafted")
  - **Characters** in older systems.

✅ **One-liner:**  
> **Tokens are the language chunks AI uses to think.**

---

## 🔹 Context Window

- The **context window** is the **limit on how much text** the model can "see" at once.
- It includes:
  - System prompts
  - User messages
  - AI responses
- Bigger context window = **longer memory** and **better conversations**.

✅ **One-liner:**  
> **The context window is the AI’s memory span.**
