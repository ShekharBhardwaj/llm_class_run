## 🛠️ Two Types of Hugging Face APIs

| Level | What It Means | When You Use It |
|:---|:---|:---|
| **Pipelines (High-Level API)** | Super simple, fast access to pre-trained models | Everyday tasks like text generation, summarization, sentiment |
| **Tokenizers & Models (Low-Level API)** | More control over tokenization, model loading, fine-tuning | Training your own models, tweaking behavior deeply |
